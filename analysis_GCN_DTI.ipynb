```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, precision_recall_curve, auc, f1_score, confusion_matrix
import pandas as pd
import seaborn as sns
```

```python
df_sage = pd.read_csv('lable_output_GCN_DTI_new.csv')
print(len(df_sage))
plt.figure(figsize=(8, 6))
for i in range(1):
    lb_cleaned = df_sage['labels'][26].strip("[]")  # Remove the square brackets
    lb = np.fromstring(lb_cleaned, dtype=float, sep=" ")
    y_true_class = (lb > 0.5).astype(int)
    # print(y_true_class)

    op_cleaned = df_sage['outputs'][26].strip("[]")  # Remove the square brackets
    op = np.fromstring(op_cleaned, dtype=float, sep=" ")
    y_pred_class = (op > 0.5).astype(int)
    # print(y_pred_class)

    cm = confusion_matrix(y_true_class, y_pred_class)

    sns.heatmap(cm, 
            annot=True,
            fmt='g', 
            xticklabels=['Pos','Neg'],
            yticklabels=['Pos','Neg'])
    

plt.ylabel('Actual', fontsize=13)
plt.title('Confusion Matrix', fontsize=17, pad=20)
plt.gca().xaxis.set_label_position('top') 
plt.xlabel('Prediction', fontsize=13)
plt.gca().xaxis.tick_top()

plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.5, 0.05, 'Prediction', ha='center', fontsize=13)
plt.show()
```

```python
from sklearn.metrics import f1_score

df_sage = pd.read_csv('lable_output_GCN_DTI_new.csv')
plt.figure(figsize=(16, 12))

thresholds = np.linspace(0, 1, 50)

for i in range( len(df_sage)):
    lb_cleaned = df_sage['labels'][i].strip("[]")  # Remove the square brackets
    lb = np.fromstring(lb_cleaned, dtype=float, sep=" ")

    op_cleaned = df_sage['outputs'][i].strip("[]")  # Remove the square brackets
    op = np.fromstring(op_cleaned, dtype=float, sep=" ")

    f1_scores = []
    for threshold in thresholds:
        y_pred_class = (op > threshold).astype(int)  # Convert probabilities to binary predictions
        f1 = f1_score(lb, y_pred_class)  # Compute F1 score
        f1_scores.append(f1)

    lb = (lb > 0.5).astype(int)
    op = (op > 0.5).astype(int)
    f1_mean = f1_score(lb, op)
    
    # Plot the F1 scores for this pair
    plt.plot(thresholds, f1_scores, label=f"Pair {i+1}: {f1_mean:.4f}")

plt.xlabel("Threshold")
plt.ylabel("F1 Score")
plt.title("F1 Score vs Threshold for Multiple Pairs")
plt.legend(loc="best")
plt.grid()
plt.show()
```
